# Outlier Removal
# Perform an outlier detection analysis on numerical variables (e.g., using the IQR method).
# Remove outliers from these features if they are not representative of typical house prices.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Get a list of all numerical column names except 'Sale Price'
num = df.select_dtypes(include=['number']).drop(columns=['Sale Price']).columns.tolist()
print(num)
```

```python
# Create boxplots for each numerical column (except 'Sale Price') to detect outliers
plt.figure(figsize=(20, 15))
for i, col in enumerate(num):
    plt.subplot(4, 4, i + 1)  # Arrange subplots in a 4x4 grid
    sns.boxplot(y=df[col])   # Boxplot for the current column
    plt.title(col)           # Set column name as the title
plt.tight_layout()           # Adjust layout to prevent overlap
plt.show()                   # Display all plots
```

```python
Q=['No of Bedrooms','No of Bathrooms','Flat Area (in Sqft)','Renovated Year','Longitude','Living Area after Renovation (in Sqft)','Lot Area after Renovation (in Sqft)','Area of the House from Basement (in Sqft)','Basement Area (in Sqft)']
```

```python
# Create a boxplot for 'Sale Price' to visualize distribution and outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Sale Price']])
plt.title('Box Plot of Sale Price')
plt.show()
```

```python
# Calculate IQR and outlier thresholds for 'Sale Price'
Q1 = df['Sale Price'].quantile(0.25)          # 25th percentile
Q3 = df['Sale Price'].quantile(0.75)          # 75th percentile
IQR = Q3 - Q1                                 # Interquartile Range
low_limit = Q1 - 1.5 * IQR                    # Lower bound for outliers
up_limit = Q3 + 1.5 * IQR                     # Upper bound for outliers
print(Q1, Q3, IQR, low_limit, up_limit)
```

```python
df['Sale Price']=df['Sale Price'].clip(low_limit,up_limit) # Cap the outliers in 'Sale Price' using the IQR method (clip values to within limits)

sns.boxplot(df['Sale Price'])# updated 'Sale Price' distribution after capping outliers
```

```python
# Calculate IQR and outlier limits for the specified column
Q1 = df[Q].quantile(0.25)
Q3 = df[Q].quantile(0.75)
IQR = Q3 - Q1
low_limit = Q1 - 1.5 * IQR
up_limit = Q3 + 1.5 * IQR
print(Q1,Q3,IQR,low_limit,up_limit)
```

```python
non_outliers = ~((df[Q] < low_limit) | (df[Q] > up_limit)).any(axis=1)
```

```python
df_new=df[non_outliers]
```
